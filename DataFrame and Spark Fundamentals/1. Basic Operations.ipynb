### Column Filtering

```python
new_df=df.select("title","imdb_rating","industry")
new_df.show(truncate=False)
```

### Row Filtering

```python
#Code generated by AI 
df_filtered = df.filter(df.release_year > 2010)
display(df_filtered)
```

```python
#Unique Industries
df_industries=df.select("industry").distinct()
display(df_industries)
#Unique Languages
df_languages=df.select("language").distinct()
display(df_languages)
```

```python
from pyspark.sql.functions import col
df_profit=df.withColumn('profit',col('revenue')-col('budget'))
another_df=df_profit.select("name","revenue","budget","profit")
display(another_df)
```

```python
##Filtering only Marvel Movies
df_marvel=df.filter(col("studio")=='Marvel Studios')
display(df_marvel)
```
